# EPLQ

## Introduction
Efficient Privacy-preserving Location-based Query (EPLQ) system allows secure querying of location data while preserving user privacy.

## Objectives
- Ensure privacy in location-based queries
- Provide role-based access (Admin/User)
- Enable secure search and retrieval

## Features
- User and Admin login system
- Encrypted queries
- Secure data storage in Firebase
- Search and filter functionality

## Tools and Technologies
- HTML, CSS, JavaScript
- Firebase Authentication & Firestore
- CryptoJS

## How to Run
1. Open `register.html` or `login.html` in a browser.
2. Register/Login using Firebase authentication.
3. Use search with passphrase for secure queries.

## Test Cases
- Verify login/logout works.
- Admin can upload POIs.
- User can search with correct passphrase.

## Conclusion
EPLQ successfully balances usability and privacy in querying sensitive location-based data.
